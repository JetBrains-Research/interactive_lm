```python
import pydriller
import os
import shutil
from tqdm.auto import tqdm

if not os.path.exists(f'tmp_repos'):
    os.mkdir(f'tmp_repos')
```

```python
counter_ = 0
if not os.path.exists(f'new_data'):
    os.mkdir(f'new_data')
if not os.path.exists(f'new_data/docs'):
    os.mkdir(f'new_data/docs')

for url in tqdm(np.unique(df_good.rep_url.values)[:3000], total=len(np.unique(df_good.rep_url.values)[:3000])):
    sub_df = df_good[df_good['rep_url'] == url]
    obj_array = []

    name = url.split('/')[-1]
    if not os.path.exists(f'new_data/docs/{name}'):
        os.mkdir(f'new_data/docs/{name}')
    else:
        continue
    if not os.path.exists(f'tmp_repos/{name}'):
        os.mkdir(f'tmp_repos/{name}')

    repo = pydriller.Repository(url, clone_repo_to=f'tmp_repos/{name}', only_modifications_with_file_types='.py',
                                num_workers=6)
    try:
        commits = list(repo.traverse_commits())
    except:
        shutil.rmtree(f'tmp_repos/{name}')
        continue

    com2hash = {c.hash:c for c in commits}
    for row_idx, row in sub_df.iterrows():
        com_hash = row['com_hash']
        com_msg = row['com_msg']
        if com_hash not in com2hash:
            continue
        commit_obj = com2hash[com_hash]

        if commit_obj.lines < 100:
            if len(commit_obj.modified_files) == 1:
                mod_file = commit_obj.modified_files[0]

                try:
                    code_before = mod_file.content_before.decode('utf-8')
                except:
                    code_before = ''
                try:
                    methods_before = mod_file.methods_before
                except:
                    methods_before = []
                names2methods_before = {el.name: el for el in methods_before}


                try:
                    code_after = mod_file.content.decode('utf-8')
                except:
                    continue
                try:
                    methods_after = mod_file.methods
                except:
                    methods_after = []
                names2methods_after = {el.name: el for el in methods_after}

                code_diff = mod_file.diff
                
                for changed_method_id, changed_method in enumerate(mod_file.changed_methods):
                    changed_method_name = changed_method.name
                    if changed_method_name in names2methods_before and changed_method_name in names2methods_after:
                        method_before = names2methods_before[changed_method_name]
                        method_after = names2methods_after[changed_method_name]

                        s_before, e_before = method_before.start_line, method_before.end_line
                        s_after, e_after = method_after.start_line, method_after.end_line

                        code_snippet_before = '\n'.join(code_before.split('\n')[(s_before-1):e_before])
                        code_snippet_after = '\n'.join(code_after.split('\n')[(s_after-1):e_after])

                        new_obj = {}
                        new_obj['code_before'] = code_snippet_before
                        new_obj['code_after'] = code_snippet_after
                        new_obj['commit_msg'] = com_msg 
                        
                        new_obj['com_py2imports_before'] = row['com_py2imports_before']
                        new_obj['com_py2imports_after'] = row['com_py2imports_after']
                        
                        new_obj['diff'] = code_diff
                        new_obj['com_hash'] = com_hash
                        obj_array.append(new_obj)
                        counter_ += 1
                
    with open(f'new_data/docs/{name}/objects.json', 'w') as f:
        json.dump(obj_array, f)
                        
    
    shutil.rmtree(f'tmp_repos/{name}')
```
